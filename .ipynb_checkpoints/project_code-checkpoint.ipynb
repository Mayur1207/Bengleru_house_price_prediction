```python
def total_sqft_clean(string):
    list1=[]
    if re.findall('[\d]+[\s]*[-]{1}[\s]*[\d]',string):
        a=re.findall('[\d]+[\s]*[-]{1}[\s]*[\d]+',string)
#         print(a[0])
        z=re.split('-',a[0])
#         print(z[0],z[1])
        a1=int(z[0])
        a2=int(z[1])
        mean=(a1+a2)/2
#         print(mean)
        list1.append(mean)
        return str(mean)
    
    
    else:
        
        if 'Acers' in string:
            list1=[]
            for i in string:
                if i.isdecimal() or i=='.':
                    list1.append(i)
            h=''.join(list1)
            Sqaure_foot=float(h)*43560
            return str(round(Square_foot),2)
        
        
        elif 'Sq. Meter' in string:
            list1=[]
            for i in string:
                if i.isdecimal() or i=='.':
                    list1.append(i)
            if list1.count('.')==2:
                list1.pop(-1)
                h=''.join(list1)
                Square_foot=float(h)*10.764
                return str(round(Square_foot,2))
            else:
                h=''.join(list1)
                Square_foot=float(h)*10.764
                return str(round(Square_foot,2))
            
            
        elif 'Sq. Yards' in string:
            list1=[]
            for i in string:
                if i.isdecimal() or i=='.':
                    list1.append(i)
                if list1.count('.')==2:
                    list1.pop(-1)
                    h=''.join(list1)
                    Square_foot=float(h)*9
                    return str(round(Square_foot,2))
                else:
                    h=''.join(list1)
                    Square_foot=float(h)*9
                    return str(round(Square_foot,2))
            
        else:
            list1=[]
            for i in string:
                if i.isdecimal() or i=='.':
                    list1.append(i)
            return ''.join(list1)
    
df['total_sqft_clean']=df['total_sqft'].apply(total_sqft_clean).astype('float')
```

# Outlier Detection and Missing Value Imputation and Outlier Handling

```python
data.describe()
```

```python
data.isna().sum()
```

```python
# size
```

```python
sns.boxplot(x=data['size'])
```

```python
# Outliers are present inn our data so we can fill missing values by using median
```

```python
data['size']=data['size'].fillna(data['size'].median())
```

```python
#Outlier handling
```

```python
q1=data['size'].quantile(0.25)
q3=data['size'].quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
mean=data.loc[data['size']<upper]['size'].mean()
mean=round(mean,2)
data.loc[data['size']>upper]=round(upper,0)
```

```python
sns.boxplot(x=data['size'])
```

```python
# Bath
```

```python
sns.boxplot(x=data['bath'])
```

```python
#There is outlier are present in bath data
```

```python
data['bath']=data['bath'].fillna(data['bath'].median())
```

```python
def outlier_handle(a):
    q1=data[a].quantile(0.25)
    q3=data[a].quantile(0.75)
    iqr=q3-q1
    lower=q1-1.5*iqr
    upper=q3+1.5*iqr
    data.loc[data[a]>upper]=round(upper,0)
    data.loc[data[a]<lower]=lower
    
```

```python
outlier_handle('bath')
```

# Model Evaluation

# Training Data Evaluation

```python
y_pred_train=linear.predict(x_train)
mse=mean_squared_error(y_train,y_pred_train)
print('mse is :',mse)
rmse=np.sqrt(mse)
print('rmse is :',rmse)
r2score=r2_score(y_train,y_pred_train)
print('r2score is :',r2score)
```

# Testing Data Evaluation

```python
y_pred_test=linear.predict(x_test)
mse=mean_squared_error(y_test,y_pred_test)
print('mse is :',mse)
rmse=np.sqrt(mse)
print('rmse is :',rmse)
r2=r2_score(y_test,y_pred_test)
print('r2 score is :',r2)
```
